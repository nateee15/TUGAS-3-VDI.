```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

# **penggunaan energi terbarukan untuk keseluruhan**

```python
df = pd.read_csv('/content/modern-renewable-energy-consumption.csv')
df
```

```python
df_asean = df[df['Entity'].isin(['Indonesia', 'Singapore', 'Malaysia','Thailand', 'Philippines','Vietnam'])]
df_asean
```

```python
df_asean = df_asean.drop(columns=['Code'])
df_asean.head(10)
```

```python
df_asean = df_asean.fillna(0)
df_asean
```

```python
# Sum the generation values for each country
df_asean['Total Generation - TWh'] = (df_asean['Other renewables (including geothermal and biomass) electricity generation - TWh'] +
                                df_asean['Solar generation - TWh'] +
                                df_asean['Wind generation - TWh'] +
                                df_asean['Hydro generation - TWh'])

# Group by country and sum the total generation
country_totals = df_asean.groupby('Entity')['Total Generation - TWh'].sum().reset_index()

# Create the bar chart using Plotly Express
fig = px.bar(country_totals, x='Entity', y='Total Generation - TWh',
             title='Total Renewable Electricity Generation by Country',
             labels={'Entity': 'Country', 'Total Generation - TWh': 'Total Electricity Generation (TWh)'},
             color='Entity')

# Show the plot
fig.show()
```

# **indonesia**

```python
df_indo = df[df['Entity'] == 'Indonesia']
df_indo.head()
```

```python
dff = df_indo.drop(columns=['Code'])
dff.head(10)
```

# **mengubah data NAN**

```python
dff_cleaned = dff.fillna(0)
dff_cleaned.head()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(dff_cleaned['Year'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'], label='Other Renewables (TWh)')
plt.plot(dff_cleaned['Year'], dff_cleaned['Solar generation - TWh'], label='Solar Generation (TWh)')
plt.plot(dff_cleaned['Year'], dff_cleaned['Wind generation - TWh'], label='Wind Generation (TWh)')
plt.plot(dff_cleaned['Year'], dff_cleaned['Hydro generation - TWh'], label='Hydro Generation (TWh)')

plt.xlabel('Year')
plt.ylabel('Electricity Generation (TWh)')
plt.title('energi terbarukan secara keseluruhan di Indonesia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(19, 8))

plt.fill_between(dff_cleaned['Year'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'], label='Other Renewables (TWh)', alpha=0.7)
plt.fill_between(dff_cleaned['Year'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'] + dff_cleaned['Solar generation - TWh'], label='Solar Generation (TWh)', alpha=0.7)
plt.fill_between(dff_cleaned['Year'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'] + dff_cleaned['Solar generation - TWh'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'] + dff_cleaned['Solar generation - TWh'] + dff_cleaned['Wind generation - TWh'], label='Wind Generation (TWh)', alpha=0.7)
plt.fill_between(dff_cleaned['Year'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'] + dff_cleaned['Solar generation - TWh'] + dff_cleaned['Wind generation - TWh'], dff_cleaned['Other renewables (including geothermal and biomass) electricity generation - TWh'] + dff_cleaned['Solar generation - TWh'] + dff_cleaned['Wind generation - TWh'] + dff_cleaned['Hydro generation - TWh'], label='Hydro Generation (TWh)', alpha=0.7)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Electricity Generation (TWh)', fontsize=14)
plt.title('energi terbarukan secara keseluruhan di Indonesia', fontsize=16)


plt.legend(fontsize=12, loc='upper left')

plt.grid(True, linestyle='--', linewidth=0.5)

plt.xticks(dff_cleaned['Year'], rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```

```python
df_per = pd.read_csv('/content/elec-fossil-nuclear-renewables.csv')
df_per.head()
```

```python
# Memfilter data untuk enam negara yang disebutkan
countries = ['Indonesia', 'Singapore', 'Malaysia', 'Thailand', 'Philippines', 'Vietnam']
df_filtered = df_per[df_per['Entity'].isin(countries)]

# Menghitung total renewables - TWh per negara
total_renewables = df_filtered.groupby('Entity')['Electricity from renewables - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)'].sum().reset_index()

# Membuat bar plot menggunakan plotly.express
fig = px.bar(total_renewables, x='Entity', y='Electricity from renewables - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)',
             title='Total Pembangkitan Listrik Terbarukan per Negara (TWh)',
             labels={'Entity': 'Negara', 'Electricity from renewables - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)': 'Total Electricity from renewables - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)'},
             color='Entity',
             color_discrete_sequence=px.colors.qualitative.Pastel)

# Menampilkan plot
fig.show()
```

```python
df_perbandingan = df_per.drop(columns='Code')
df_perbandingan = df_perbandingan[df_perbandingan['Entity'] == 'Indonesia']

df_perbandingan.columns = ['Entity', 'Year', 'renewables - TWh', 'nuclear - TWh', 'fossil fuels - TWh']

df_perbandingan
```

```python
fig = px.line(df_perbandingan, x='Year', y=['renewables - TWh', 'nuclear - TWh', 'fossil fuels - TWh'],
              labels={'value': 'Million (Ton)', 'variable': 'Energy Source'},
              title='perbandingan energi terbarukan, nuclear dan fosil ')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
fig = px.area(df_perbandingan, x='Year', y=['nuclear - TWh', 'renewables - TWh', 'fossil fuels - TWh'],
              labels={'value': 'Million (Ton)', 'variable': 'Energy Source'},
              title='perbandingan energi terbarukan, nuclear dan fosil ')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
df_pie = df_perbandingan.melt(id_vars='Year', var_name='Energy Source', value_name='Million (Ton)')

fig = px.pie(df_pie, names='Energy Source', values='Million (Ton)', title='Perbandingan Energi Terbarukan, Nuclear dan Fosil')
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(width=700, height=700)

fig.show()
```
